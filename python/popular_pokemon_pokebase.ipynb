```python
import pandas as pd

pokemons = []
df = pd.DataFrame()
for i in range(1, 151):
    pokemon = pb.pokemon(i)
    index = 0
    pokemon_habilidades = []
    pokemon_habilidades_dano = []
    for move in pokemon.moves:
        if index < 3:
            if(move.move.power != None):
                pokemon_habilidades.append(move.move.name)
                pokemon_habilidades_dano.append(move.move.power)
                index += 1
            
        else:
            break
    try:
        pokemons.append([pokemon.name, pokemon.height, pokemon.weight, pokemon.id, pokemon.stats[0].base_stat,
            pokemon.stats[1].base_stat, pokemon.stats[2].base_stat, pokemon.stats[3].base_stat, 
            pokemon.stats[4].base_stat, pokemon.stats[5].base_stat, pokemon_habilidades[0],
            pokemon_habilidades_dano[0], pokemon.types[0].type, pokemon.types[1].type,
            pokemon_habilidades, pokemon_habilidades_dano])
    except: 
        pass

df = pd.DataFrame(pokemons, columns=['nome', 'altura', 'peso', 'id', 'hp', 'ataque', 'defesa', 'ataque_especial', 'defesa_especial', 'velocidade', 'nome_primeira_habilidade', 'dano_primeira_habilidade', 'tipo1', 'tipo2', 'habilidades', 'habilidades_dano'])
df
```
